<h2><center><font color='red'>Importing Libraries</font></h2>

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('darkgrid')
%matplotlib inline
```

<h2><center><font color='red'>Loading the Dataset</font></h2>

```python
dataset = sns.load_dataset('titanic')
dataset.drop(columns=['who', 'adult_male', 'alone', 'alive', 'deck', 'embark_town', 'pclass', 'sibsp'], inplace=True)
dataset['embarked'].fillna(dataset['embarked'].mode()[0], inplace=True)
print(dataset.head())
```

<h2><center><font color='red'>Distribution Plots</font></h2>

```python
sns.displot(dataset.fare, color='black', facecolor='red', alpha=0.75, kde=True, bins=45)
plt.xticks(np.arange(0, 525, 25), rotation=90)
plt.yticks(np.arange(0, 375, 25))
plt.xlabel("\nFare")
plt.ylabel("")
plt.title("Distribution Plot for Fare\n")
plt.show()
```

```python
sns.jointplot(x=dataset.age, y=dataset.fare, kind='hex', color='darkblue')
plt.xlabel("\nAge")
plt.ylabel("Fare")
plt.title("Joint Plot for Age vs. Fare\n")
plt.show()
```

```python
plt.figure(figsize=(6, 4))

sns.rugplot(dataset.fare, color='darkgreen')
plt.xlabel("\nFare")
plt.ylabel("")
plt.title("Rug Plot for Fare\n")
plt.show()
```

<h2><center><font color='red'>Categorical Plots</font></h2>

```python
sns.barplot(x=dataset.sex, y=dataset.fare, palette='seismic')
plt.xlabel("\nGender")
plt.ylabel("Fare")
plt.title("Bar Plot for Gender vs. Fare\n")
plt.show()
```

```python
sns.countplot(x='survived', data=dataset, palette='twilight')
plt.xlabel("\nSurvived")
plt.yticks(np.arange(0, 600, 50))
plt.title("Count Plot for checking number of dead and alive people\n")
plt.show()
```

```python
sns.boxplot(x=dataset['class'], y=dataset['fare'], palette='Spectral_r')
plt.yticks(np.arange(0, 600, 50))
plt.title("Box Plot for Class vs. Fare\n")
plt.show()
```

```python
sns.violinplot(x=dataset['class'], y=dataset['fare'], palette='rocket')
plt.yticks(np.arange(0, 600, 50))
plt.title("Violin Plot for Class vs. Fare\n")
plt.show()
```

<h2><center><font color='red'>Advanced Plots</font></h2>

```python
sns.stripplot(x=dataset.sex, y=dataset.fare, hue=dataset.survived, palette='viridis', s=3.5)
plt.xlabel("\nGender")
plt.ylabel("Fare")
plt.title("Strip Plot for Gender vs. Fare\n")
plt.show()
```

```python
sns.swarmplot(x=dataset['class'], y=dataset.fare, hue=dataset.sex, palette='inferno_r', s=3.5)
plt.ylabel("Fare")
plt.title("Swarm Plot for Class vs. Fare\n")
plt.show()
```

<h2><center><font color='red'>Matrix Plot</font></h2>

```python
dataset['sex'] = dataset['sex'].map({'male':0, 'female':1})
dataset['embarked'] = dataset['embarked'].map({'S':0, 'C':1, 'Q':2})
dataset['class'] = dataset['class'].map({'First':0, 'Second':1, 'Third':2})
```

```python
sns.heatmap(round(dataset.corr(), 2), annot=True, cbar=False, cmap='Blues', square=True, annot_kws={'size':'8', 'color':'black'})
plt.show()
```
