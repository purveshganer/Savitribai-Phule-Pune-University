```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

sns.set_style('darkgrid')
%matplotlib inline
```

<h2><center><font color='red'>Loading the Dataset</font></h2>

```python
dataset = sns.load_dataset('iris')
print(dataset.head())
```

<h2><center><font color='red'>Displaying Features and their Datatypes</font></h2>

```python
dataset.info()
```

<h2><center><font color='red'>Data Visualization</font></h2>

```python
columns = dataset.columns[:-1]

counter = 1

plt.figure(figsize=(9, 9))
plt.subplots_adjust(wspace=0.75, hspace=0.55)

for c in columns:

    plt.subplot(2, 2, counter)
    sns.histplot(dataset[c], color=np.random.rand(3, ))
    plt.title(f"Histogram for {c[:5].capitalize()+' '+c[6:].capitalize()}\n")
    plt.xlabel(f"{c[:5].capitalize()+' '+c[6:].capitalize()}")

    counter += 1

plt.show()

plt.show()
```

```python
counter = 1

plt.figure(figsize=(9, 9))
plt.subplots_adjust(wspace=0.75, hspace=0.45)

for c in columns:

    plt.subplot(2, 2, counter)
    sns.boxplot(dataset[c], color=np.random.rand(3, ))
    plt.title(f"Boxplot for {c[:5].capitalize()+' '+c[6:].capitalize()}\n")
    plt.xlabel(f"{c[:5].capitalize()+' '+c[6:].capitalize()}")
    counter += 1

plt.show()
```
