<h2><center><font color='red'>Importing Libraries</font></h2>

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

sns.set_style('darkgrid')
%matplotlib inline
```

<h2><center><font color='red'>Loading the Dataset</font></h2>

```python
dataset = sns.load_dataset('titanic')
print(dataset.head())
```

<h2><center><font color='red'>Box Plot for Gender vs. Age</font></h2>()

```python
sns.boxplot(y=dataset.age, x=dataset.sex, hue=dataset.survived, palette='seismic')
plt.yticks(np.arange(0, 85, 5))
plt.show()
```
